–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π, —á—Ç–æ–±—ã –∫–∞–∂–¥–∞—è —Ñ–∏—à–∫–∞ –ø–æ—Å–µ—Ç–∏–ª–∞ —Ö–æ—Ç—è –±—ã —Ä–∞–∑ –ø–æ–∑–∏—Ü–∏—é, –∫–æ—Ç–æ—Ä—É—é –≤—ã–±—Ä–∞–ª –¥–ª—è –Ω–µ–µ –ü–µ—Ç—è.

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –≤—ã–≤–µ–¥–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü–∏–π. –î–ª—è –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è –æ–ø–µ—Ä–∞—Ü–∏–π –≤–ª–µ–≤–æ, –≤–ø—Ä–∞–≤–æ, –≤–Ω–∏–∑, –≤–≤–µ—Ä—Ö –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–∏–º–≤–æ–ª—ã ùêø,ùëÖ,ùê∑,ùëà —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

```python
import math
def find_path(s_pos: list[int], e_pos: list[int]):
    start = s_pos.copy()
    finish = e_pos.copy()
    result = []
    while start != finish:
        if finish[0] > start[0]:
            start[0] += 1
            result += "D"
        elif finish[0] < start[0]:
            start[0] -= 1
            result += "U"
        if finish[1] > start[1]:
            start[1] += 1
            result += "R"
        elif finish[1] < start[1]:
            start[1] -= 1
            result += "L"
    return result

def solve(s_pos: list[list[int]], e_pos: list[list[int]],n,m):
    result = ""
    while len(result) < 2 * m * n and len(s_pos) > 0:
        tmp = list(map(''.join, map(sorted, map(find_path, s_pos, e_pos))))
        for move in tmp[tmp.index(max(tmp))]:
            match move:
                case "D":
                    for el in s_pos: el[0] += 1 if math.fabs(el[0] + 1) <= n else 0
                case "U":
                    for el in s_pos: el[0] -= 1 if math.fabs(el[0] - 1) <= n else 0
                case "R":
                    for el in s_pos: el[1] += 1 if math.fabs(el[1] + 1) <= n else 0
                case "L":
                    for el in s_pos: el[1] -= 1 if math.fabs(el[1] - 1) <= n else 0
        result += tmp[tmp.index(max(tmp))]
        for i in range(len(s_pos)-1,-1,-1):
            if s_pos[i] == e_pos[i]: s_pos.pop(i); e_pos.pop(i)
    if len(s_pos) > 0:
        return -1
    return str(len(result)) + "\n" + result

n, m, k = map(int,input().strip().split())
s_pos = [list(map(int,input().strip().split())) for _ in range(k)]
e_pos = [list(map(int,input().split())) for _ in range(k)]
print(solve(s_pos,e_pos,n,m))
```

–ó–∞—Ö–≤–∞—Ç–∏—Ç—å –∫–æ—Ä–æ–ª–µ–≤—Å—Ç–≤–æ –≤ —Ç–æ—á–∫–µ 30, —Å—Ç–æ–∏–º–æ—Å—Ç—å 3‚ãÖ(30‚àí5)=75.\
–°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å 3+6+12+24+3+48+75=171. –ù–µ–ª—å–∑—è –ø–æ–ª—É—á–∏—Ç—å –º–µ–Ω—å—à—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å.


```python
result = []
for i in range(int(input())):
    n,a,b = map(int, input().strip().split())
    vh = list(map(int, input().strip().split()))
    sm = sum(vh)
    rs=sm * b
    s=0
    for j in range(n):
        s += vh[j]
        rs = min(rs, (a + b)*vh[j] + (sm-s - (n-j-1)*vh[j])*b)
 
    result.append(rs)
print(*(el for el in result), end='')
```

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –Ω–µ –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ–ø–µ—Ä–∞—Ü–∏—é. 13‚ãÖ2+37‚ãÖ4=174.

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç—å –ø–æ–¥–º–∞—Å—Å–∏–≤ [3,5]. –¢–æ–≥–¥–∞ ùëé=[1,8,3,6,7,6] –∏ 1‚ãÖ5+8‚ãÖ9+3‚ãÖ6+6‚ãÖ8+7‚ãÖ8+6‚ãÖ6=235.


```python
n = int(input())
a = list(map(int, input().strip().split()))
b = list(map(int, input().strip().split()))
mx = p = sum(a[i] * b[i] for i in range(n))
def solve(l, r, d=p):
    global mx
    while l >= 0 and r < n:
        d -= (a[l] - a[r]) * (b[l] - b[r])
        mx = max(mx, d)
        l -= 1
        r += 1 

        
for i in range(n):
    solve(i-1,i+1)
    solve(i,i+1)
print(mx)
```
