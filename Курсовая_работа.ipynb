```python
def solve(t: int,n: int, k: int, z: int, a: list[int]) -> int:
    ans = 0
    s = 0
    mx = 0
    for t in range(z + 1):
        pos = k - 2 * t
        if pos < 0:
            continue
        mx = 0
        s = 0
        for i in range(pos + 1):
            if i < n - 1:
                mx = max(mx, a[i] + a[i + 1])
            s += a[i]
        ans = max(ans, s + mx * t)
    return max(ans, s + mx * t)


def inp(n: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            final_error = mess_error
            result[:] = map(int, input(mess).strip().split())
            if len(result) != n:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                if number_of_input == 3:
                    final_error += str(" ".join([str(i + 1) for i in range(n)]))
                raise BaseException
            if number_of_input == 1 and not (1 <= result[0] <= 10 ** 4):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ —á–∏—Å–ª–∞ t
                final_error = "–ü–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: (1 ‚â§ t ‚â§ 10^4)"
                raise BaseException
            elif number_of_input == 2 \
                    and (not (2 <= result[0] < 10 ** 5)
                         or not (1 <= result[1] <= result[0] - 1)
                         or not (0 <= result[2] <= min(5, result[1]))):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ –¥–ª—è —á–∏—Å–µ–ª n, k, z
                final_error = ""
                if not (2 <= result[0] < 10 ** 5):
                    final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 2 ‚â§ n ‚â§ 10^5"
                if not (1 <= result[1] <= result[0] - 1):
                    final_error += "\n" if final_error != "" else ""
                    final_error += "–î–∏–∞–ø–∞–∑–æ–Ω k: 1 ‚â§ k ‚â§ n-1"
                if not (0 <= result[2] <= min(5, result[1])):
                    final_error += "\n" if final_error != "" else ""
                    final_error += "–î–∏–∞–ø–∞–∑–æ–Ω z: 0 ‚â§ Z ‚â§ min(5,k)"
                raise BaseException
            elif number_of_input == 3 and \
                    (max(result) > 10 ** 4 or
                     min(result) < 1):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –Ω—É–ª—è –≤ –º–∞—Å—Å–∏–≤–µ
                final_error = "–î–∏–∞–ø–∞–∑–æ–Ω ai: 1 ‚â§ ai ‚â§ 10^4"
                raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result


if __name__ == '__main__':
    result = []
    t = inp(1, "–í–≤–µ–¥–∏—Ç–µ t: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1", 1)[0]
    for _ in range(t):
        n, k, z = inp(3, "–í–≤–µ–¥–∏—Ç–µ n, k, z (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª): ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 3 2 1", 2)
        a = inp(n, "–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ a: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: ", 3)
        result.append(solve(t,n, k, z, a))
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\n"+"\n".join(str(el) for el in result))
```

```python
def solve(n: int, a: list[int]):
    result = ""
    result += "1 1\n"+str((n-1)*a[0])+"\n"
    if n>=2:
      result += str(2)+ " " + str(n) + "\n"
      result += " ".join([str(int((n-1)*a[i])) for i in range(1,n)]) + "\n"
    else:result +=("1 1\n0")
    result += str(1) + " " + str(n) + "\n"
    result += " ".join([str(int(-n*a[i])) for i in range(n)]) + "\n"
    return result
def inp(count: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            final_error = mess_error
            result[:] = map(int, input(mess).strip().split())
            if len(result) != count:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                raise BaseException
            if number_of_input == 1 and not (1 <= result[0] <= 10 ** 4):
                final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 1 ‚â§ n ‚â§ 10^4"
                raise BaseException
            elif number_of_input == 2 and (min(result) < -109 or max(result) > 109):
                final_error += "–î–∏–∞–ø–∞–∑–æ–Ω a: -109 ‚â§ a ‚â§ 109"
                raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result


if __name__ == "__main__":
    n = inp(1, "–í–≤–µ–¥–∏—Ç–µ n: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1", 1)[0]
    a = inp(n, "–í–≤–µ–¥–∏—Ç–µ –∞, b: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2 3 4", 2)
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\n" + str(solve(n,a)))
```

–°—É–º–º–∞—Ä–Ω–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å 3+6+12+24+3+48+75=171. –ù–µ–ª—å–∑—è –ø–æ–ª—É—á–∏—Ç—å –º–µ–Ω—å—à—É—é —Å—Ç–æ–∏–º–æ—Å—Ç—å.



```python
def solve(n: int, a: int, b: int, vh: list[int]) -> int:
    sm = sum(vh)
    rs = sm * b
    s = 0
    for j in range(n):
        s += vh[j]
        rs = min(rs, (a + b) * vh[j] + (sm - s - (n - j - 1) * vh[j]) * b)
    return rs

def inp(n: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            # t (1‚â§t‚â§10 ^ 3)
            # n, a –∏ b (1‚â§n‚â§2‚ãÖ105; 1‚â§a,b‚â§10^5).
            # (1‚â§xn‚â§108).
            final_error = mess_error
            if number_of_input == 3:
                final_error += str(" ".join([str(i + 1) for i in range(n)]))
            result[:] = map(int, input(mess).strip().split())
            if len(result) != n:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                raise BaseException
            if number_of_input == 1 and not (1 <= result[0] <= 10 ** 3):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ —á–∏—Å–ª–∞ t
                final_error = "–ü–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: (1 ‚â§ t ‚â§ 10^3)"
                raise BaseException
            elif number_of_input == 2 \
                    and (not (2 <= result[0] < 2 * 10 ** 5)
                         or not (1 <= result[1] <= 10 ** 5)
                         or not (1 <= result[2] <= 10 ** 5)):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ –¥–ª—è —á–∏—Å–µ–ª n, k, z
                final_error = ""
                if not (2 <= result[0] < 2 * 10 ** 5):
                   final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 1 ‚â§ n ‚â§ 2 * 10^5"
                if not (1 <= result[1] <= 10 ** 5):
                    final_error += "\n" if final_error != "" else ""
                    final_error += "–î–∏–∞–ø–∞–∑–æ–Ω a: 1 ‚â§ a ‚â§ 10 ** 5"
                if not (1 <= result[2] <= 10 ** 5):
                    final_error += "\n" if final_error != "" else ""
                    final_error += "–î–∏–∞–ø–∞–∑–æ–Ω a: 1 ‚â§ a ‚â§ 10 ** 5"
                raise BaseException
            elif number_of_input == 3 and \
                    (max(result) > 108 or
                     min(result) < 1):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –Ω—É–ª—è –≤ –º–∞—Å—Å–∏–≤–µ
                final_error = "–î–∏–∞–ø–∞–∑–æ–Ω xi: 1 ‚â§ xi ‚â§ 108"
                raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result


if __name__ == "__main__":
    result = []
    t = inp(1, "–í–≤–µ–¥–∏—Ç–µ t: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 6", 1)[0]
    for i in range(t):
        n, a, b = inp(3, "–í–≤–µ–¥–∏—Ç–µ n, a, b (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª): ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 3 2 1", 2)
        vh = inp(n, "–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ a: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: ", 3)
        result.append(solve(n,a,b,vh))
    print("\n".join(str(el) for el in result))
```

### –ó–∞–¥–∞—á–∞ 7

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–í–æ –≤—Ç–æ—Ä–æ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëé1,‚Ä¶,ùëéùëõ (1‚â§ùëéùëñ‚â§1018).

–í —Ç—Ä–µ—Ç—å–µ–π —Å—Ç—Ä–æ–∫–µ –∑–∞–ø–∏—Å–∞–Ω–æ ùëö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëè1,‚Ä¶,ùëèùëö (1‚â§ùëèùëó‚â§1018).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ ùëö —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª. ùëó-–µ –∏–∑ —ç—Ç–∏—Ö —á–∏—Å–µ–ª –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ä–∞–≤–Ω–æ –ù–û–î(ùëé1+ùëèùëó,‚Ä¶,ùëéùëõ+ùëèùëó).

–ü—Ä–∏–º–µ—Ä\

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
2 3 8 24


```python
from math import*

def gcd(x, y):
    if(y == 0):
        return x
    else:
        return gcd(y, x % y)

def solve(a: list[int], b: list[int]) -> list[int]:
    s = 0
    for i in a[1:]:
        s = gcd(s, a[0] - i)
    result = []
    for i in b:
        result.append(gcd(s, a[0] + i))
    return result


def inp(count: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            final_error = mess_error
            result[:] = map(int, input(mess).strip().split())
            if len(result) != count:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                raise BaseException
                if number_of_input == 1 \
                    and not (1 <= result[0] <= 2 * 10 ** 5)\
                        or not (1 <= result[1] <= 2 * 10 ** 5):
                    final_error = ""
                    if not (1 <= result[0] < 10 ** 5):
                        final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 1 ‚â§ n ‚â§ 10^5"
                    if not (1 <= result[1] <= 10 ** 5):
                        final_error += "\n" if final_error != "" else ""
                        final_error += "–î–∏–∞–ø–∞–∑–æ–Ω m: 1 ‚â§ m ‚â§ 10^5"
                    raise BaseException
            elif number_of_input in range(2,4)\
                    and (not (1 <= result[0] <= 10 ** 18) \
                      or not (1 <= result[1] <= 10 ** 18)):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ –¥–ª—è —á–∏—Å–µ–ª sxi, syi, fxi, fyi
                litera = "a" if number_of_input == 2 else "b"
                if (min(result) < 1 or max(result) > 10 ** 19):
                    final_error = "–î–∏–∞–ø–∞–∑–æ–Ω "+litera+"i: 1 ‚â§ " + litera + "i ‚â§ 10^18"
                raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result


if __name__ == "__main__":
    n, m = inp(2, "–í–≤–µ–¥–∏—Ç–µ n, m (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª): ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2", 1)
    a = inp(n, "–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ a (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª): ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2 3", 2)
    b = inp(m, "–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ b (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª): ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2 3", 3)
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\n"+"\n".join(str(el) for el in solve(a,b)))
```

```python
def solve(l, r):
    global p, mx
    d=p
    while l >= 0 and r < n:
        d -= (a[l] - a[r]) * (b[l] - b[r])
        mx = max(mx, d)
        l -= 1
        r += 1


def inp(count: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            final_error = mess_error
            result[:] = map(int, input(mess).strip().split())
            if len(result) != count:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                if number_of_input in range(2,4):
                    final_error += str(" ".join([str(i + 1) for i in range(n)]))
                raise BaseException
            if number_of_input == 1 and not (1 <= result[0] <= 5 * 10 ** 3):
                final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 1 ‚â§ n ‚â§ 5000"
                raise BaseException
            elif number_of_input in range(2,4)\
                    and (not (1 <= result[0] <= 10 ** 7) \
                      or not (1 <= result[1] <= 10 ** 7)):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ –¥–ª—è —á–∏—Å–µ–ª sxi, syi, fxi, fyi
                litera = "a" if number_of_input == 2 else "b"
                if (min(result) < 1 or max(result) > 10 ** 7):
                    final_error = "–î–∏–∞–ø–∞–∑–æ–Ω "+litera+"i: 1 ‚â§ " + litera + "i ‚â§ 10^7"
                raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result


if __name__ == "__main__":
    result = []
    for _ in range(int(input())):
        n = inp(1, "–í–≤–µ–¥–∏—Ç–µ n: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1", 1)[0]
        a = inp(n, "–í–≤–µ–¥–∏—Ç–µ a: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2 3", 2)
        b = inp(n, "–í–≤–µ–¥–∏—Ç–µ b: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2 3", 3)
        mx = p = sum(a[i] * b[i] for i in range(n))
        for i in range(n):
            solve(i - 1, i + 1)
            solve(i, i + 1)
        result.append(mx)
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\n"+"\n".join(str(el) for el in result))
```

```python
from math import fabs


def solve(a: int,b: int,n: int) -> str:
    if a == 0:
        return  0 if b == 0 else 'No solution'
    elif b % a:
       return 'No solution'
    else:
        for x in range(-1000, 1001):
            if x ** n == b // a:
                return x
        return 'No solution'

    
def inp(count: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            final_error = mess_error
            result[:] = map(int, input(mess).strip().split())
            if len(result) != count:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                raise BaseException
                if number_of_input == 1 and\
                        (not (fabs(result[0]) <= 1000) or\
                         not (fabs(result[1]) <= 1000) or
                         not (1 <= result[2] <= 10)):
                    #  A, B, n, —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏—Ö —É—Å–ª–æ–≤–∏—è–º: |A|,|B|‚â§1000, 1‚â§n‚â§10.
                    final_error = ""
                    if not (fabs(result[0]) <= 1000):
                        final_error += "–î–∏–∞–ø–∞–∑–æ–Ω a: |A| ‚â§ 1000"
                    if not (fabs(result[1]) <= 1000):
                        final_error += "\n" if final_error != "" else ""
                        final_error += "–î–∏–∞–ø–∞–∑–æ–Ω b: |B| ‚â§ 1000"
                    if not (1 <= result[2] <= 10):
                        final_error += "\n" if final_error != "" else ""
                        final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 1 ‚â§ n ‚â§ 10"
                    raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result


if __name__ == "__main__":
    result = []
    for _ in range(int(input())):
        a, b, n =  inp(3, "–í–≤–µ–¥–∏—Ç–µ a, b, n (—á–µ—Ä–µ–∑ –ø—Ä–æ–±–µ–ª): ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1 2 3", 1)
        result.append(solve(a,b,n))
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\n" + "\n".join([str(el) for el in result]))
```

```python
from collections import Counter


def solve(Sum: int) -> bool:
    if d[Sum] > 0:
        d[Sum] -= 1
        return True
    if Sum == 1: return False
    return solve(Sum // 2) and solve((Sum + 1) // 2)


def inp(count: int, mess: str, mess_error: str, number_of_input: int):
    """ –ú–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ –¥–∞–Ω–Ω—ã—Ö. –ü—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –≤–≤–æ–¥–∏–º—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤,
    —Å–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –≤–≤–æ–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ–± –æ—à–∏–±–∫–µ –∏ –Ω–æ–º–µ—Ä –≤–≤–æ–¥–∞. """
    result = []
    while True:
        try:
            final_error = mess_error
            result[:] = map(int, input(mess).strip().split())
            if len(result) != count:
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–≤–µ–¥–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
                if number_of_input == 3:
                    final_error += str(" ".join([str(i + 1) for i in range(count)]))
                raise BaseException
            if number_of_input == 1 and not (1 <= result[0] <= 10 ** 4):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ —á–∏—Å–ª–∞ t
                final_error = "–ü–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: (1 ‚â§ t ‚â§ 10^4)"
                raise BaseException
            elif number_of_input == 2 and not (1 <= result[0] < 2 * 10 ** 4):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –≤–≤–æ–¥–∞ –¥–ª—è —á–∏—Å–µ–ª n, k, z
                final_error += "–î–∏–∞–ø–∞–∑–æ–Ω n: 2 ‚â§ n ‚â§ 2 * 10^5"
                raise BaseException
            elif number_of_input == 3 and \
                    (max(result) > 10 ** 9 or
                     min(result) < 1):
                # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –Ω—É–ª—è –≤ –º–∞—Å—Å–∏–≤–µ
                final_error = "–î–∏–∞–ø–∞–∑–æ–Ω ai: 1 ‚â§ ai ‚â§ 10^9"
                raise BaseException
            break  # –ï—Å–ª–∏ –≤–≤–æ–¥–∏–º—ã–µ –¥–∞–Ω–Ω—ã–µ —è–≤–ª—è—é—Ç—Å—è –≤–µ—Ä–Ω—ã–º–∏ - –≤—ã—Ö–æ–¥ –∏–∑ —Ü–∏–∫–ª–∞
        except BaseException:
            print(final_error)
    return result
"""–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùë° (1‚â§ùë°‚â§104) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö.
–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (1‚â§ùëõ‚â§2‚ãÖ105).
–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç ùëõ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª ùëé1,ùëé2,‚Ä¶,ùëéùëõ (1‚â§ùëéùëñ‚â§109)."""


if __name__ == "__main__":
    global d
    result = []
    t = inp(1, "–í–≤–µ–¥–∏—Ç–µ t: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1", 1)[0]
    for _ in range(t):
        n = inp(1, "–í–≤–µ–¥–∏—Ç–µ n: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: 1", 2)[0]
        a = inp(n, "–í–≤–µ–¥–∏—Ç–µ –º–∞—Å—Å–∏–≤ a: ", "–í–≤–µ–¥–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ —Å–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∏–º–µ—Ä—É: ", 3)
        d = Counter(a)
        result.append("YES" if solve(sum(a)) else "NO")
    print("\n".join(str(el) for el in result))
```
